```python
import numpy as np
from scipy import linalg
from math import sin,exp
import matplotlib.pylab as plt

def function(x):
    return sin(x/5.)*exp(x/10.)+5.*exp(-x/2.)

X=np.arange(1.,15.1,0.1)
Y=np.array([function(x) for x in X])
#print X
#print Y


```

```python
## polynomial 1 degree

A1=np.empty((0,2))
Y1=np.array([1,15])
#print arr
for i in range(2):
    A_row=list()
    for j in range(2):
        A_row.append(Y1[i]**j)
    A1=np.append(A1,np.array([A_row]),axis=0)
B1=np.array([function(x) for x in Y1])

#print A
#print B
    
X1=linalg.solve(A1,B1)
#print X1

def appfunc1(x):return X[0]+X[1]*x
YbyFunc1=np.array([appfunc1(x) for x in X])
```

```python
%matplotlib inline
```

```python
plt.plot(X,Y,X,YbyFunc1)
```

# Polynomial2 degree

```python
A2=np.empty((0,3))
Y2=np.array([1,8,15])
for i in range(3):
    A_row=list()
    for j in range(3):
        A_row.append(Y2[i]**j)
    A2=np.append(A2,np.array([A_row]),axis=0)
B2=np.array([function(x) for x in Y2])

#print A2
#print B2

X2=linalg.solve(A2,B2)
print X2

def appfunc2(x):return X2[0]+X2[1]*x+X2[2]*(x**2)
Ybyfunc2=np.array([appfunc2(x) for x in X])

plt.plot(X,Y,X,Ybyfunc2)
plt.show()

```

# polynomiam 3 degree

```python
Y3=np.array([1,4,10,15])
A3=np.empty((0,4))
for i in range(4):
    A_row=list()
    for j in range(4):
        A_row.append(Y3[i]**j)
    A3=np.append(A3,np.array([A_row]),axis=0)
B3=np.array([function(x) for x in Y3])
#print A3
#print B3

X3=linalg.solve(A3,B3)
#print X3

def appfunc3(x):return X3[0]+X3[1]*x+X3[2]*(x**2)+X3[3]*(x**3)
Ybyfunc3=np.array([appfunc3(x) for x in X])
plt.plot(X,Y,X,Ybyfunc3)
plt.show()
```
